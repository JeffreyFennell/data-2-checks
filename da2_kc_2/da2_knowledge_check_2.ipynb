Reformat all columns so that everything is uniform and snake case.

```python
fixed_columns = {'ScoreRecent':'Score_Recent', 
                 'GradeRecent':'Grade_Recent', 
                 'DateRecent':'Date_Recent', 
                 'Score2':'Score_2', 
                 'Grade2':'Grade_2', 
                 'Date2':'Date_2',
                 'Score3':'Score_3', 
                 'Grade3':'Grade_3', 
                 'Date3':'Date_3', 
                 'permit_number':'Permit_Number', 
                 'facility_type':'Facility_Type',
                 'facility_type_description':'Facility_Type_Description', 
                 'subtype':'Subtype', 
                 'subtype_description':'Subtype_Description',
                 'premise_name':'Premise_Name', 
                 'premise_address':'Premise_Address', 
                 'premise_city':'Premise_City', 
                 'premise_state':'Premise_State',
                 'premise_zip':'Premise_Zip', 
                 'opening_date':'Opening_Date', 
                 'ObjectId': 'Object_Id'}
```

```python
df.rename(columns= fixed_columns, inplace= True)
```

Change columns comtaining date data to datetime datatypes.

```python
df[['Date_Recent','Date_2','Date_3','Opening_Date']] = df[['Date_Recent','Date_2','Date_3','Opening_Date']].apply(pd.to_datetime)
```
